```python
df=data.copy()
```

```python
df.head()
```

```python
df["date_index"] = df["Date_month_year"] + " " + df["Time"]
df["date_index"]
```

```python
df.head()
```

```python
df["date_index"] = pd.to_datetime(df["date_index"])
df["Date_month_year"] = pd.to_datetime(df["Date_month_year"])
```

```python
df = df.set_index(df["date_index"])
df.head()
```

```python
df.Date = pd.to_datetime(df.date_index, format='%Y/%m/%d %I:%M:%S %p')
df.index = pd.DatetimeIndex(df.Date)
```

```python
plt.figure(figsize=(11,5))
df.resample('M').size().plot(legend=False)
plt.title('Number of complaints per month ')
plt.xlabel('Months')
plt.ylabel('Number of complaints')
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(11,4))
df.resample('D').size().rolling(10).sum().plot()
plt.title('Rolling sum of all complaints')
plt.ylabel('Number of complaints')
plt.xlabel('Days')
plt.show()
```
